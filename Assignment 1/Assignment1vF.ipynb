```python
from sklearn.preprocessing import MinMaxScaler
norm = df[['Price','Landsize']].to_numpy()
scaler = MinMaxScaler()
normalized_data = scaler.fit_transform(norm)
print(normalized_data)
```

```python
normalized_df = pd.DataFrame(normalized_data, columns=['Nor_Price','Nor_Landsize'])
print(normalized_df)
```

```python
from sklearn.preprocessing import StandardScaler
normz = df[['Price', 'Landsize']]
scaler = StandardScaler()
standardized_data = scaler.fit_transform(normz)
print(standardized_data)
```

```python
standardized_df = pd.DataFrame(standardized_data, columns=['St_Price', 'St_Landsize'])
print(standardized_df)
```
